Outcomes:
- Check how the consumers consume it:
Messages go to random partitions when key is not mentioned in message. In group-2 , there is only 1 consumer so all partitions are assigned to it.





















- How partitions are alloted to consumers: 
Partitions assigned to consumers randomly, since in group-2 there is only 1 consumer, all partitions are assigned to it





- How reallocation happens if a consumer is added:  Partitions get redistributed within the consumers in the consumer group.



- How reallocation happens if a consumer is removed: Partitions are redistributed, in this case all partitions get mapped to consumer remaining
